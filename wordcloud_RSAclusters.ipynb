```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df = pd.read_csv('decode_related.csv')
keys=df.Terms
Terms=keys.values.tolist()
Correlation=df.Correlation.values.tolist()
from collections import Counter
dictionary = Counter(df)
dictionary={}

for i in range(25):
    dictionary[Terms[i]]=Correlation[i]
    
wordcloud = WordCloud(max_font_size=50, max_words=200, background_color="white").generate_from_frequencies(dictionary)
import matplotlib.pyplot as plt
plt.imshow(wordcloud,interpolation='bilinear')
#plt.figure(figsize=(16,12))
plt.axis('off')

#wordcloud.to_file("first_review.png",dpi=300)
plt.savefig('Final_Plot_related_elmo_w2.png', dpi=300, transparent=False, bbox_inches='tight')
plt.show()
```
